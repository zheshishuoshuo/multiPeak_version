```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from emcee.backends import HDFBackend

def plot_posterior_sns_pairplot(
    hdf5_file,
    discard=1000,
    thin=10,
    labels=None,
    truths=None,
    savefile=None
):
    """
    从 emcee 的 HDF5 后端文件中读取样本并用 seaborn 绘制 pairplot 图。

    参数：
    - hdf5_file: str，HDF5 文件路径
    - discard: int，burn-in 步数
    - thin: int，抽样间隔
    - labels: list[str]，变量标签（列名）
    - truths: list[float]，真值（可选）
    - savefile: str，保存文件名（可选）
    """
    backend = HDFBackend(hdf5_file, read_only=True)
    samples = backend.get_chain(discard=discard, thin=thin, flat=True)

    ndim = samples.shape[1]
    if labels is None:
        labels = [f"$\\theta_{i}$" for i in range(ndim)]
    print(np.shape(samples))
    # 构建 pandas DataFrame
    df = pd.DataFrame(samples, columns=labels)

    # pairplot 绘图
    g = sns.pairplot(
        df,
        corner=True,           # 不画对角线右上部分
        diag_kind="hist",      # 直方图对角线
        plot_kws=dict(s=5, alpha=0.2),  # 控制散点图
    )

    if truths is not None:
        for i, truth in enumerate(truths):
            if truth is None:
                continue
            var = labels[i]
            # 对角线添加真值线
            ax = g.axes[i][i]
            ax.axvline(truth, color="red", linestyle="--")

    if savefile:
        plt.savefig(savefile, bbox_inches="tight", dpi=300)
        print(f"[INFO] 图已保存到: {savefile}")

    plt.show()

plot_posterior_sns_pairplot(
    hdf5_file="/home/shuoz/sl_inference/mcmc_chain_no_cache.h5",
    discard=1000,
    thin=10,
    labels=[r"$\mu_{\rm DM}$", r"$\sigma_{\rm DM}$", r"$\mu_\alpha$", r"$\sigma_\alpha$"],
 #    truths=(12.73729339602048,
 # 0.38500126475763946,
 # 0.07255019362558475,
 # 0.07476414612714799),
    # truths =[12.7, 0.34, 0.1, 0.13],
    savefile="posterior_pairplot_eta.png"
)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from emcee.backends import HDFBackend

def plot_marginal_histograms(
    hdf5_file,
    discard=1000,
    thin=10,
    labels=None,
    truths=None,
    bins=100,
    savefile=None
):
    """
    绘制每个维度的边际后验分布 histogram，用于排查多峰结构。

    参数：
    - hdf5_file: str，HDF5 路径
    - discard: int，丢弃前 burn-in 步数
    - thin: int，抽样间隔
    - labels: list[str]，参数标签
    - truths: list[float]，真值位置（可选）
    - bins: int，直方图的箱数
    - savefile: str，保存文件名（可选）
    """
    backend = HDFBackend(hdf5_file, read_only=True)
    samples = backend.get_chain(discard=discard, thin=thin, flat=True)

    ndim = samples.shape[1]
    if labels is None:
        labels = [f"$\\theta_{i}$" for i in range(ndim)]

    fig, axes = plt.subplots(ndim, 1, figsize=(6, 1.8 * ndim), constrained_layout=True, dpi=200)

    for i in range(ndim):
        ax = axes[i]
        ax.hist(samples[:, i], bins=bins, density=True, alpha=0.7, color='steelblue', edgecolor='black')
        ax.set_xlabel(labels[i])
        ax.set_ylabel("PDF")
        if truths is not None and truths[i] is not None:
            ax.axvline(truths[i], color="red", linestyle="--", lw=2)

    if savefile:
        plt.savefig(savefile, dpi=300)
        print(f"[INFO] 图已保存到: {savefile}")

    plt.show()

# 示例调用
plot_marginal_histograms(
    hdf5_file="/home/shuoz/sl_inference/mcmc_chain_no_cache.h5",
    discard=1000,
    thin=10,
    labels=[r"$\mu_{\rm DM}$", r"$\sigma_{\rm DM}$", r"$\mu_\alpha$", r"$\sigma_\alpha$"],
    # truths=[12.73, 0.32, 0.1, 0.13],
    bins=80,
    savefile="posterior_marginals_hist.png"
)
```

```python
import numpy as np
import pandas as pd
from scipy.interpolate import RegularGridInterpolator
import matplotlib.pyplot as plt

# === 插值器构建函数 ===
def load_A_phys_interpolator(filename):
    df = pd.read_csv(filename)
    muDM_unique = np.sort(df['mu_DM'].unique())
    sigmaDM_unique = np.sort(df['sigma_DM'].unique())
    A_grid = df.pivot(index='mu_DM', columns='sigma_DM', values='A_phys').values
    return RegularGridInterpolator((muDM_unique, sigmaDM_unique), A_grid,
                                   bounds_error=False, fill_value=None)

# 加载两个不同精度的插值器
interp_low = load_A_phys_interpolator("A_phys_table100.csv")
interp_high = load_A_phys_interpolator("A_phys_table1000.csv")

# === 在 mu_DM ∈ [12.6, 13.0]，sigma_DM 固定测试 ===
muDM_vals = np.linspace(12.6, 13.0, 100)
sigma_DM_test = 0.3  # 可调（也可以多选几个不同值一起测）

# === 计算两个插值器的结果 ===
pts = np.column_stack([muDM_vals, [sigma_DM_test] * len(muDM_vals)])
A_low = interp_low(pts)
A_high = interp_high(pts)

# === 误差分析 ===
abs_error = np.abs(A_low - A_high)
rel_error = np.abs((A_low - A_high) / A_high)

# === 打印最大/平均误差 ===
print(f"最大绝对误差: {np.max(abs_error):.3e}")
print(f"平均绝对误差: {np.mean(abs_error):.3e}")
print(f"最大相对误差: {np.max(rel_error)*100:.2f}%")
print(f"平均相对误差: {np.mean(rel_error)*100:.2f}%")

# === 可视化 ===
plt.figure(figsize=(8, 5))
plt.plot(muDM_vals, A_low, label='Low Res')
plt.plot(muDM_vals, A_high, '--', label='High Res')
plt.fill_between(muDM_vals, A_low, A_high, color='gray', alpha=0.3, label='Difference')
plt.xlabel(r'$\mu_{\rm DM}$')
plt.ylabel(r'$A_{\rm phys}$')
plt.legend()
plt.title(f'Comparison at ')
plt.savefig('error.png', bbox_inches="tight", dpi=300)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import RegularGridInterpolator
from emcee.backends import HDFBackend

# === 加载插值器 ===
def load_A_phys_interpolator(filename):
    df = pd.read_csv(filename)
    muDM_unique = np.sort(df['mu_DM'].unique())
    sigmaDM_unique = np.sort(df['sigma_DM'].unique())
    A_grid = df.pivot(index='mu_DM', columns='sigma_DM', values='A_phys').values
    return RegularGridInterpolator((muDM_unique, sigmaDM_unique), A_grid,
                                   bounds_error=False, fill_value=None)

interp_low = load_A_phys_interpolator("A_phys_table100.csv")
interp_high = load_A_phys_interpolator("A_phys_table1000.csv")

# === 后验样本（只取 mu_DM 维度） ===
backend = HDFBackend("/home/shuoz/sl_inference/mcmc_chain1000.h5", read_only=True)
samples = backend.get_chain(discard=2000, thin=1, flat=True)
muDM_samples = samples[:, 0]  # 假设第 0 维是 mu_DM

# === 设置评估区间与插值 ===
muDM_vals = np.linspace(12.6, 13.0, 200)
sigma_DM_test = 0.3
pts = np.column_stack([muDM_vals, [sigma_DM_test] * len(muDM_vals)])
A_low = interp_low(pts)
A_high = interp_high(pts)

# === 左轴：联合后验的直方图 ===
fig, ax1 = plt.subplots(figsize=(8, 5))

# 只选取指定区间的样本用于直方图
mask = (muDM_samples > 12.6) & (muDM_samples < 13.0)
hist_vals, bins, _ = ax1.hist(muDM_samples[mask], bins=50, alpha=0.6, color='gray', label=r'Posterior of $\mu_{\rm DM}$')

ax1.set_xlabel(r'$\mu_{\rm DM}$')
ax1.set_ylabel('Posterior Density', color='gray')
ax1.tick_params(axis='y', labelcolor='gray')
ax1.grid(True)
ax1.legend(loc='upper left')

# === 右轴：A_phys 插值线 ===
ax2 = ax1.twinx()
ax2.plot(muDM_vals, A_low, label='A_phys (Low Res)', color='tab:blue')
ax2.plot(muDM_vals, A_high, '--', label='A_phys (High Res)', color='tab:orange')
ax2.set_ylabel(r'$A_{\rm phys}$', color='black')
ax2.tick_params(axis='y', labelcolor='black')
ax2.legend(loc='upper right')

# === 保存 & 显示 ===
plt.title(f'Posterior of $\mu_{{DM}}$ and $A_{{phys}}$ at $\sigma_{{DM}}={sigma_DM_test}$')
plt.tight_layout()
plt.savefig("posterior_vs_interpolator.png", dpi=300)
plt.show()
```
